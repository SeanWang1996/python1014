```python
import pandas as pd
import numpy as np

pd1 = pd.DataFrame(np.random.randn(6, 7),
                   index=list(range(0, 12, 2)),
                   columns=list(range(0, 7, 1)))
pd1
```

```python
pd1.index = pd1.iloc[:,0]
pd1
```

```python
pd1.sort_index()
```

```python
pd2 = pd.DataFrame(np.random.randn(6, 7),
                   index=list(range(0, 12, 2)),
                   columns=list(range(0, 7, 1)))
pd2
```

```python
pd2.columns = pd2.iloc[1,:]
pd2
```

```python
pd2.sort_index(axis=1)
```

```python
pd3 = pd.DataFrame(np.random.randn(6, 7),
                   index=list(range(0, 12, 2)),
                   columns=list(range(0, 7, 1)))
pd3
```

```python
pd3.sort_values(by=0)
```

```python
s1 = pd.Series([4,-15,7,7,2,2,0,0,4])
s1.rank()
```

```python
s1.rank(method='first')
```

```python
s1.rank(method='max')
```

```python
s1.rank(method='min')
```
